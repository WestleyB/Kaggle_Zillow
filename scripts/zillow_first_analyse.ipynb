```python

col_tmp = []
for el in df_train.columns.tolist():
    if df_train[el].dtypes == 'float64':
        col_tmp.append(el)

df_train[col_tmp] = df_train[col_tmp].fillna(0)
df_train[col_tmp].head()
```

```python
col_tmp = []
for el in df_train.columns.tolist():
    if df_train[el].dtypes == 'object':
        col_tmp.append(el)
        df_train[el]=df_train[el].fillna(1)
        lbl = LabelEncoder()
        lbl.fit(list(df_train[el].values))
        df_train[el] = lbl.transform(list(df_train[el].values))
        
df_train[col_tmp].head()
```
